## Extract image size information from Image URLs

```python
imgs= df['ImgUrl']
imgs.dropna(inplace=True)
imgs.reset_index(drop=True, inplace=True)
imgs
```

```python
imgs_list = imgs.to_list()
```

```python
#find image size: 
import urllib
from PIL import ImageFile

def getsizes(uri):
    # get file size *and* image size (None if not known)
    file = urllib.request.urlopen(uri)
    size = file.headers.get("content-length")
    if size: size = int(size)
    p = ImageFile.Parser()
    while 1:
        data = file.read(1024)
        if not data:
            break
        p.feed(data)
        if p.image:
            return size, p.image.size
            break
    file.close()
    return size, None

img_size= getsizes('https://m.media-amazon.com/images/I/71X2BLG8q-L._AC_UL320_.jpg')
```

```python
img_size[1]
```

```python
x=[]
for i in range(len(imgs_list)):
    img_size= getsizes(imgs_list[i])
    x.append(img_size[1])
print(x)
```

```python
#drop na columns 
df2= df[df['ImgUrl'].notna()]
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df2["ImgSize"] = x
df2.head()
```

```python
df2.to_csv('df_with_imgsize.csv')
```

### Add Image height, width, Area column and tidy up data: to integer values for analysis

```python
df3 = pd.read_csv('df_with_imgsize.csv')
df3['ImgSize']=df3['ImgSize'].str.replace('(','')
df3['ImgSize']=df3['ImgSize'].str.replace(')','')
```

```python
df3.head()
```

```python
df3['ImgHeight']= df3['ImgSize'].str.slice(0, 3, 1)
```

```python
df3['ImgWidth']= df3['ImgSize'].str.slice(4, 8, 1)
```

```python
df3['ImgHeight']=df3['ImgHeight'].str.replace(' ','')
```

```python
df3['ImgWidth']=df3['ImgWidth'].str.replace(' ','')
```

```python
df3['ImgWidth']=df3['ImgWidth'].apply(pd.to_numeric,errors='coerce')
df3['ImgHeight']=df3['ImgHeight'].apply(pd.to_numeric,errors='coerce')
df3
```

```python
df3['ImgArea']= df3['ImgHeight'] * df3['ImgWidth']
```

```python
df3.dropna(inplace=True)
df3.reset_index(drop=True, inplace=True)
df3
```

```python
df3.drop(df3.columns[0], axis = 1, inplace = True)
df3
```

```python
df3['Rating']= df3['Rating'].str.slice(0, 3, 1)
df3['ReviewCount']=df3['ReviewCount'].str.replace(',','')
df3['Price']=df3['Price'].str.replace('$','')
df3['ReviewCount']=df3['ReviewCount'].apply(pd.to_numeric,errors='coerce')
df3['Price']=df3['Price'].apply(pd.to_numeric,errors='coerce'
df3['Rating']=df3['Rating'].apply(pd.to_numeric,errors='coerce'
```
